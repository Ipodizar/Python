```python
# 한글 폰트 설정
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 생존유무  pie 그래프
f,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].set_title('생존확률')
titanic.Survived.value_counts().plot.pie(autopct='%1.1f%%',shadow=True,explode = [0,0.1],ax=ax[0])

ax[1].set_title('생존인원')
titanic.Survived.value_counts().plot(kind='bar',ax=ax[1])
plt.show()
```

```python
# 선실등급별 생존유무
titanic.groupby('Pclass')[['Survived']].sum()
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'],margins=True).style.background_gradient()
```

```python
# seaborn
import seaborn as sns
```

```python
sns.countplot(data=titanic, x='Pclass', hue= 'Survived', palette=['red','blue'], legend=True) # legend : 범례 표시 여부
plt.show()
```

```python
sns.countplot(data=titanic, x='Gender', hue= 'Survived') # legend : 범례 표시 여부
plt.show()
```

```python
# 선실등급별, 성별에 따라서 생존확률이 높다는 것을 알 수 있음
# 두 특성을 한꺼번에 비교하는 시각화 툴
# catplot
sns.catplot(data=titanic, x='Pclass', y='Survived', hue='Gender', kind='point', height = 4, aspect = 1.5)
```

```python
from cProfile import label


sns.kdeplot(
    titanic[titanic.Survived == 1], x='Age', label = 'Survived == 1', fill=True
)

sns.kdeplot(
    titanic[titanic.Survived == 0], x='Age', label = 'Survived == 0', fill=True
)

plt.legend(loc = 'best')
```

```python
# 위의 표를 응용해서 Pclss별로 생존유무에 대한 시각화
f, ax = plt.subplots(1, 3, figsize=(12, 5))
ax[0].set_title('1등석')
ax[1].set_title('2등석')
ax[2].set_title('3등석')
sns.kdeplot(
    titanic[titanic.Pclass == 1], x='Age', hue='Survived', label = 'Pclass== 1', fill=True, ax=ax[0]
)

sns.kdeplot(
    titanic[titanic.Pclass == 2], x='Age', hue='Survived', label = 'Pclass== 2', fill=True, ax=ax[1]
)

sns.kdeplot(
    titanic[titanic.Pclass == 3], x='Age', hue='Survived', label = 'Pclass== 3', fill=True, ax=ax[2]
)

plt.legend(loc = 'best')
```

```python
# 생존자 중에서 나이가 어린 사람이 많다.
```

```python
# 클래스가 높을수록 나이 많은 사람의 비중이 커짐
# 나이대가 변하면서 생존률이 어떻게 될까
```

```python
# 나이대의 범위를 점점 넓혀가면서 생존률의 변화
plt.figure(figsize=(10, 5))
multi_age_range = [titanic[titanic.Age<i].Survived.mean() for i in range(1, 80)]
plt.plot(range(1, 80), multi_age_range)
plt.show()
```
