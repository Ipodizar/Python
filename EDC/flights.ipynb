```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.info()
# category : 문자열 - 범위형 데이터
```

```python
flights.T
```

```python
flights.pivot(index = 'year', columns = 'month', values = 'passengers')
# .pivot(index='', columns='', values='')
```

```python
# 연도별 승객수의 변화 lineplot - sns
import seaborn as sns
import matplotlib.pylab as plt
# sns.lineplot(data=flights, x = 'year', y = 'passengers', errorbar = ('ci', 95))
sns.lineplot(data=flights, x = 'year', y = 'passengers', hue = 'month', errorbar = None)
# sns 시각화는 신뢰구간도같이 표시하는 경우
# 데이터가 변동될 수 있는 구간
# 신뢰구간이 넓으면 불확실성이 커짐 (데이터 분산이 크거나 샘플 수가 작음)
# 신뢰구간이 좁으면 불확실성이 안정적 (데이터가 모여있거나 샘플 수가 많음)
# plt.legend(loc='best') default 
plt.show()
```
